# Maxwell-Boltzmann Distribution

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline

def plot_distribution(x,y,xlabel,ylabel):
    """Plot the x and y values of a distribution."""
    # Plotting parameters
    fgs = (12,8)     # figure size
    lw = 8           # line width
    lfs = 18         # label font size
    tfs = 16         # tick font size
    xlp,ylp = 30,50  # label padding
    ylr = 0          # y-axis label rotation
    # Generate the plot
    fig,ax = plt.subplots(figsize=fgs)
    plt.plot(x,y,linewidth=lw)
    plt.tick_params(labelsize=tfs)
    plt.xlabel(xlabel,fontsize=lfs,labelpad=xlp)
    plt.ylabel(ylabel,fontsize=lfs,rotation=ylr,labelpad=ylp)
    plt.tight_layout()
    return fig,ax
```

```python
# Function parameters
npoints = 1001

m = 1
k = 8.6173303e-5 # eV/K
T = 293          # K
```

## Energy

```python
E = np.linspace(0,0.2,npoints)
M_E = 2*np.sqrt(E/np.pi)*(1/(k*T))**(3/2)*np.exp(-E/(k*T))
```

```python
# Plot the Maxwell-Boltzmann distribution for energy
figE,axE = plot_distribution(E,M_E,'$E$','$M(E)$')
# Plot a vertical line at the most probably energy, E = k*T
E_p = k*T*np.ones(200)
y = np.arange(0,20,0.1)
axE.plot(0.5*E_p,y,'g')
#plt.savefig('mb-dist_E.png')
```

## Speed

```python
v = np.linspace(0,0.8,npoints)
M_v = 4*np.pi*v**2*(m/(2*np.pi*k*T))**(3/2)*np.exp(-m*v**2/(2*k*T))
```

```python
# Plot the Maxwell-Boltzmann distribution for speed (blue)
figv,axv = plot_distribution(v,M_v,'$v$','$M(v)$')
# Overlay the distribution for speed converted to energy (orange)
E_v = 0.5*m*v**2
axv.plot(E_v,M_v,linewidth=8)
# Plot a vertical line at the most probably energy, E = k*T
axv.plot(E_p[:40],y[:40],'g')
#plt.savefig('mb-dist_v.png')
```
