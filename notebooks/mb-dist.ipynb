# Maxwell-Boltzmann Distribution

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
# Plotting parameters
fgs = (12,8)      # figure size
lw = 8           # line width
lfs = 18         # label font size
tfs = 16         # tick font size
xlp,ylp = 30,50  # label padding
ylr = 0          # y-axis label rotation
# Function parameters
npoints = 1001
```

```python
m = 1
k = 8.6173303e-5 # ev/K
T = 293          # K
```

## Speed

```python
v = np.linspace(0,0.8,npoints)
M_v = 4*np.pi*v**2*(m/(2*np.pi*k*T))**(3/2)*np.exp(-m*v**2/(2*k*T))
```

```python
figv,axv = plt.subplots(figsize=fgs)
plt.plot(v,M_v,linewidth=lw)
plt.tick_params(labelsize=tfs)
plt.xlabel('$v$',fontsize=lfs,labelpad=xlp)
plt.ylabel('$M(v)$',fontsize=lfs,rotation=ylr,labelpad=ylp)
plt.tight_layout()
plt.savefig('mb-dist_v.png')
```

## Energy

```python
E = np.linspace(0,0.2,npoints)
M_E = 2*np.sqrt(E/np.pi)*(1/(k*T))**(3/2)*np.exp(-E/(k*T))
```

```python
figE,axE = plt.subplots(figsize=fgs)
plt.plot(E,M_E,linewidth=lw)
plt.tick_params(labelsize=tfs)
plt.xlabel('$E$',fontsize=lfs,labelpad=xlp)
plt.ylabel('$M(E)$',fontsize=lfs,rotation=ylr,labelpad=ylp)
plt.tight_layout()
plt.savefig('mb-dist_E.png')
```
